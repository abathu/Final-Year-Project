```python
# Reading the TSV file
df4_1 = pd.read_csv(df4_path1, sep='\t')
df4_2 = pd.read_csv(df4_path2, sep='\t')
df4_3 = pd.read_csv(df4_path3, sep='\t')

display(df4_1 ,df4_2 , df4_3 )
print(df4_1['labels'].value_counts())
```

```python
# drop counter speech. it is hard to say what category is
df4_1 = df4_1[df4_1['labels'] != 'CounterSpeech']
df4_2 = df4_2[df4_2['labels'] != 'CounterSpeech']
df4_3 = df4_3[df4_3['labels'] != 'CounterSpeech']
```

```python
def assign_values(row):
    if row['labels'] == 'Neutral':
        return pd.Series([0, 0])
    elif row['labels'] == 'IdentityDirectedAbuse':
        return pd.Series([1, 0])
    elif row['labels'] == 'AffiliationDirectedAbuse':
        return pd.Series([1, 0])
    elif row['labels'] == 'PersonDirectedAbuse':
        return pd.Series([0, 1])
    elif row['labels'] == 'IdentityDirectedAbuse,PersonDirectedAbuse':
        return pd.Series([1, 1])
    elif row['labels'] == 'AffiliationDirectedAbuse,IdentityDirectedAbuse':
        return pd.Series([1, 0])
    elif row['labels'] == 'AffiliationDirectedAbuse,PersonDirectedAbuse':
        return pd.Series([1, 1])
    elif row['labels'] == 'AffiliationDirectedAbuse,IdentityDirectedAbuse,PersonDirectedAbuse':
        return pd.Series([1, 1])
    elif row['labels'] == 'CounterSpeech,PersonDirectedAbuse':
        return pd.Series([0, 1])
    elif row['labels'] == 'CounterSpeech,IdentityDirectedAbuse':
        return pd.Series([1, 0])
    elif row['labels'] == 'AffiliationDirectedAbuse,CounterSpeech':
        return pd.Series([1, 1])
    else:
        return pd.Series([0, 0])  # or handle other cases as needed

df4_1[['hate speech', 'offensive language']] = df4_1.apply(assign_values, axis=1)
df4_2[['hate speech', 'offensive language']] = df4_2.apply(assign_values, axis=1)
df4_3[['hate speech', 'offensive language']] = df4_3.apply(assign_values, axis=1)

```

```python
df4_1.drop(['id','labels'], axis=1, inplace=True)
df4_2.drop(['id','labels'], axis=1, inplace=True)
df4_3.drop(['id','labels'], axis=1, inplace=True)
df4 = pd.concat([df4_1, df4_2,df4_3], ignore_index=True)
display(df4)
```

```python
df4.to_csv('df4.csv', quotechar='"', escapechar='\\', index=False)  
```

```python
def balance_features(df,feature_name,remove_percentage = 0.5 , remove_value =0 ):
    # Step 1: Filter the DataFrame
    condition_df = df[df[feature_name] == remove_value]

    # Step 2: Sample rows to remove
    rows_to_remove = condition_df.sample(frac=remove_percentage)

    # Step 3: Drop these rows from the original DataFrame
    df = df.drop(rows_to_remove.index)

    return df
```

```python
train_dfs = []
test_dfs = []
```

```python
import re 
mention_pattern = re.compile(r'@\S+')
symbol_pattern = re.compile(r'[\&\[\]\(\)®]')
gt_pattern = re.compile(r'&gt;')
tweet_reply_pattern = re.compile(r'RT')

def remove_urls_mentions_symbols(text):

    # reutrn None if not is string 
    # drop later 
    if not isinstance(text, str):
        return None

    text = text.replace('\n', ' ')
    text = mention_pattern.sub('', text)
    text = gt_pattern.sub('', text)
    text = symbol_pattern.sub('', text)
    text = tweet_reply_pattern.sub('', text)

    return text
```

```python
import pandas as pd

# List of file names
file_names = ['df1.csv','df2.csv','df4.csv', 'df5.csv']

column_names = ['text', 'hate speech', 'offensive language']  

dfs = [pd.read_csv(file, header=0, names=column_names) for file in file_names]

for dataframe in dfs:
    dataframe['text'] = dataframe['text'].apply(remove_urls_mentions_symbols)
    dataframe["toxic"] = dataframe[['hate speech','offensive language']].max(axis=1)
    distribution_normalized = dataframe['toxic'].value_counts(normalize=True)

    print(distribution_normalized)


    display (dataframe)

```

```python
### remove null value 
for dataframe in dfs:
    print (dataframe.isnull().sum())
    dataframe = dataframe.dropna()
    print (dataframe.isnull().sum())
```

```python
## make data distribution more balanced
## def balance_features(df,feature_name,remove_percentage = 0.5 , remove_value =0 ):
dfs[1] = balance_features(dfs[1],"toxic",0.9, 0)
dfs[2] = balance_features(dfs[2],"toxic",0.8, 0)
```

```python
### split dataframe to train and test 

from sklearn.model_selection import train_test_split
for dataframe in dfs:
    df_train , df_test = train_test_split( dataframe, train_size=0.6, random_state=42)
    train_dfs.append(df_train)
    test_dfs.append(df_test)
```

```python
train_final = pd.concat(train_dfs, ignore_index=True)
```
